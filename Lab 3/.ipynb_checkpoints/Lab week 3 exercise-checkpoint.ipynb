# 1. Classification

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

import numpy as np
import matplotlib.pyplot as plt
```

```python
digits = datasets.load_digits()

X = digits.data
y = digits.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
train_scores = []
test_scores = []

for i in range(1, 9):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    train_scores.append(knn.score(X_train, y_train))
    test_scores.append(knn.score(X_test, y_test))

print(train_scores)
print(test_scores)
```

```python
plt.plot(range(1,9), train_scores, label='Training scores')
plt.plot(range(1,9), test_scores, label='Test scores')
plt.xlabel('number of neighbors')
plt.ylabel('score')
plt.legend()
```

Based on the score results plotted above, k=4 provides the most accurate results and is therefore the best value for k.

```python
from sklearn.metrics import classification_report
from sklearn.svm import SVC
```

```python
linear_svc = SVC(kernel='linear')
poly_svc = SVC(kernel='poly')
sigmoid_svc = SVC(kernel='sigmoid')
```

```python
linear_svc.fit(X_train, y_train)
y_pred = linear_svc.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
poly_svc.fit(X_train, y_train)
y_pred = poly_svc.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
sigmoid_svc.fit(X_train, y_train)
y_pred = sigmoid_svc.predict(X_test)
print(classification_report(y_test, y_pred))
```

Based on the classification reports above, the linear and poly kernel svc models are better classifiers than the sigmoid kernel svc model. The linear kernel appears to be slightly better at classifying the digits based on the f1-scores for each digit.

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
```

```python
classifier_scores = []
classifiers = ['K Nearest Neighbour', 'SVM', 'Gaussian Naïve Bayes']

#refit the KNN classifier for k=4
knn = KNeighborsClassifier(n_neighbors = 4) # k=4 yielded the hightst accuracy
knn.fit(X_train, y_train)

classifier_scores.append(knn.score(X_test, y_test))
classifier_scores.append(linear_svc.score(X_test, y_test))
classifier_scores.append(gnb.score(X_test, y_test))
```

```python
plt.bar(classifiers, classifier_scores)
plt.xlabel('classifier')
plt.ylabel('accuracy')
```

# 2. Regression

```python
import pandas as pd
```

```python
life_expectancy_data = pd.read_csv('gapminder.csv')
display(life_expectancy_data.shape)
display(life_expectancy_data.head(3))
```

```python
sns.set(rc = {'figure.figsize':(15,8)}) #make the figure larger and legible 
sns.heatmap(life_expectancy_data.corr(), annot=True, cmap="RdYlGn")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression

lin_reg = LinearRegression()
log_reg = LogisticRegression(max_iter=1000)
```

```python
plt.scatter(life_expectancy_data["fertility"], life_expectancy_data["life"])
```

```python
X = life_expectancy_data.iloc[:, 1].values.reshape(-1, 1) # converts fertility feature into a numpy array
y = life_expectancy_data.iloc[:, 7].values.reshape(-1, 1) # converts life feature into a numpy array

lin_reg.fit(X, y)
lin_reg.score(X, y)
```

```python
y_pred = lin_reg.predict(X)

plt.scatter(X, y)
plt.plot(X, y_pred, ls='--', color='red')
```

```python
# remove non-numeric feature and the variable we are trying to predict
trimmed_life_expectancy_data = life_expectancy_data.drop(columns=['life', 'Region'])
```

```python
lin_reg.fit(trimmed_life_expectancy_data, y)
lin_reg.score(trimmed_life_expectancy_data, y)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# apply 5-fold cross validation to both linear regressions
display(cross_val_score(lin_reg, X, y))
display(cross_val_score(lin_reg, trimmed_life_expectancy_data, y))
```

```python
credit_card_data = pd.read_csv('creditcard.csv')
display(credit_card_data.shape)
display(credit_card_data.head(3))
```

```python
frauds = credit_card_data.loc[credit_card_data['Class'] == 1]
non_frauds = credit_card_data.loc[credit_card_data['Class'] == 0]
print(len(frauds), 'frauds, ', len(non_frauds), 'non-frauds')
```

```python
sns.scatterplot(x=credit_card_data['Amount'], y=credit_card_data['Class'], hue=credit_card_data['Class'])
```

Based on the scatter plot above it is apparent that fradulent transactions generally are of a lower amount. Higher amount transactions are much less likely to be fradulent.

```python
# remove the variable we are trying to predict
trimmed_credit_card_data = credit_card_data.drop(columns='Class')

X = trimmed_credit_card_data
y = credit_card_data['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

log_reg.fit(X_train, y_train)
log_reg.score(X_test, y_test)
```

```python
y_pred = log_reg.predict(X_test)
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
```
