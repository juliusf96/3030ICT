In this exercise, you will be asked to explore time series data in several ways.

You will be working with time series of Open Power System Data ([OPSD](https://open-power-system-data.org/)) for Germany. 

The data set includes daily electricity consumption, wind power production, and solar power production between 2006 and 2017. 
```
- Date — The date (yyyy-mm-dd format)
- Consumption — Electricity consumption in GWh
- Wind — Wind power production in GWh
- Solar — Solar power production in GWh
- Wind+Solar — Sum of wind and solar power production in GWh
```

# Data Exploration

````python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
````

````python
# Load file
df = pd.read_csv('opsd_germany_daily.csv')
df.tail(5)
````

````python
# Description
df.describe()
````

Then, make sure to transform the dates into datetime format in pandas :

````python
# Transform the dates into datetime format
df['Date'] = pd.to_datetime(df['Date'])

type(df.index)
````

### 1. Plotting and understanding a time series

Q1 - Plot the different time series. What do you understand, what do you notice ?

````python
plt.figure(figsize=(12,8))
plt.plot(df['Date'], df['Consumption'], label='Consumption', linewidth=0.5)
plt.plot(df['Date'], df['Wind+Solar'], label='Production', linewidth=0.5, color='orange')
plt.title('Consumption vs. Production')
plt.legend()
plt.show()
````

Observations:
- Consumption is higher than production
- Production (data?) begins in 2012
- There is a seasonality component, both to the consumption and the production 
- Consumption appears to be stable over time (flat trend)
- Production appears to be increasing over time (positive trend)
- There appears to be more noise in the production
- Fluctuations in the production increase over time

We can also avoid plotting lines and simply plot dots for the consumption :

````python
plt.figure(figsize=(12,8))
plt.scatter(df['Date'], df['Consumption'], s=10)
plt.title('Consumption over time')
plt.show()
````

And plot the production over time :

````python
plt.figure(figsize=(12,8))
plt.scatter(df['Date'], df['Wind'], s=10, label='Wind')
plt.plot(df['Date'], df['Solar'], label='Solar', linewidth=0.5, color='orange')
plt.title('Energy production over time')
plt.legend()
plt.show()
````

What can we notice ?
+ the consumption is larger than the production
+ we have no overall production data before 2012
+ the production and the consumption are both seasonal, i.e. they vary within a year given some pattern
+ there is a trend in production (the higher part), and outliers downwards

How do consumption and production compare ?
+ We notice that wind based production increases over time, and is rather high in winter (for example more storms).
+ On the other hand, the solar production is higher in summer (seems logical)
+ The consumption is higher in winter

````python
import datetime

plt.figure(figsize=(12,8))
plt.plot(df['Date'], df['Consumption'], label='Consumption', linewidth=0.5)
plt.plot(df['Date'], df['Wind'], label='Wind', linewidth=0.5, color='orange')
plt.plot(df['Date'], df['Solar'], label='Solar', linewidth=0.5, color='green')
plt.title('Consumption vs. Production')
plt.xlim(datetime.date(2011, 11, 30), datetime.date(2018, 2, 1))
plt.legend()
plt.show()
````

### 2. Change the scale

Q2 - Try to change the scale to understand the recurrent patterns !

````python
# Plot over the year
plt.figure(figsize=(12,8))
plt.plot(df['Date'], df['Consumption'], label='Consumption', linewidth=0.5)
plt.title('Electricity consumption in 2017')
plt.xlim(datetime.date(2017, 1, 1), datetime.date(2018, 1, 1))
plt.show()
````

It confirms the trend: more consumption in winter time, and less in summer. We can then zoom on a given month :

````python
# Plot over a month
plt.figure(figsize=(12,8))
plt.plot(df['Date'], df['Consumption'], label='Consumption', linewidth=0.5)
plt.title('Electricity consumption in July 2017')
plt.xlim(datetime.date(2017, 7, 1), datetime.date(2017, 8, 1))
plt.ylim(1000, 1450)
plt.show()
````

When zooming on a month, we oberve another seasonality. There seems to be days of the week in which people consume less. For example, here's what happens when we highlight Sundays :

````python
# Plot on Sundays
sunday_dates = [datetime.date(2017, 7, 2)]

while sunday_dates[-1] < datetime.date(2017, 7, 31):
    sunday_dates.append(sunday_dates[-1] + datetime.timedelta(days=7))
    

plt.figure(figsize=(12,8))
plt.plot(df['Date'], df['Consumption'], label='Consumption', linewidth=0.5)
plt.title('Electricity consumption in July 2017')
plt.xlim(datetime.date(2017, 7, 1), datetime.date(2017, 8, 1))
plt.ylim(1000, 1450)
plt.vlines(sunday_dates, 1000, 1450, color='red')
plt.show()
````

### 3. Further exploration

Q3 - To further explore the dataset, we might need to group the data. Create new columns for the consumption (by year and month), and create box-plots with seaborn. Comment on your results.

````python
# new columns 'year' and 'month'
df_2 = df.reset_index()
df_2['year'] = pd.DatetimeIndex(df['Date']).year
df_2['month'] = df['Date'].dt.month_name().str.slice(stop=3)
years = df_2['year'].unique()
print(years)
print(df_2.dtypes)
df_2.head()
````

````python
# Boxplot year
import seaborn as sns
plt.figure(figsize=(12,8))
sns.boxplot(x="year", y="Consumption", data=df_2)
plt.title('Aggregated values by year')
plt.show()
````

We see that the consumption remains fairly constant over years between 2006 and 2017. 

````python
# Boxplot month
plt.figure(figsize=(12,8))
sns.boxplot(x="month", y="Consumption", data=df_2)
plt.title('Aggregated values by month')
plt.show()
````

We see that the electricity consumption is generally lower in the summer months and higher in the winter months.
